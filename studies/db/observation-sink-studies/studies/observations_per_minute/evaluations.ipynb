```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import datetime
import json

with open('primary_signal_observations_per_minute_timestamp_counts.json', 'r') as fp:
    primary_signal_observations_per_minute_timestamp_counts = json.load(fp)

dates = []
counts = []

for timestamp, count in primary_signal_observations_per_minute_timestamp_counts.items():
    timestamp = float(timestamp)
    if timestamp < 1694607780:
        continue
    dates.append(datetime.datetime.fromtimestamp(timestamp))
    counts.append(count)

fig, ax = plt.subplots(figsize=(20,10))
ax.plot(dates, counts, label="Total")
ax.xaxis.set_major_formatter(mdates.DateFormatter('%m-%d %H:%M'))
ax.set_xlabel('Timestamp')
ax.set_ylabel('Count')
ax.set_title('Count of total primary signal observations per minute')
ax.legend()
plt.show()
```
