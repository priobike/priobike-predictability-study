```python
import pandas as pd
import matplotlib.pyplot as plt
import json

# Read the data
calendar_df = pd.read_csv('hvv_rohdaten_gtfs/calendar.txt')

service_ids_monday = calendar_df[calendar_df['monday'] == 1]['service_id'].values
service_ids_tuesday = calendar_df[calendar_df['tuesday'] == 1]['service_id'].values
service_ids_wednesday = calendar_df[calendar_df['wednesday'] == 1]['service_id'].values
service_ids_thursday = calendar_df[calendar_df['thursday'] == 1]['service_id'].values
service_ids_friday = calendar_df[calendar_df['friday'] == 1]['service_id'].values
service_ids_saturday = calendar_df[calendar_df['saturday'] == 1]['service_id'].values
service_ids_sunday = calendar_df[calendar_df['sunday'] == 1]['service_id'].values

trips_df = pd.read_csv('hvv_rohdaten_gtfs/trips.txt')

trips_df_monday = trips_df[trips_df['service_id'].isin(service_ids_monday)]
trips_df_tuesday = trips_df[trips_df['service_id'].isin(service_ids_tuesday)]
trips_df_wednesday = trips_df[trips_df['service_id'].isin(service_ids_wednesday)]
trips_df_thursday = trips_df[trips_df['service_id'].isin(service_ids_thursday)]
trips_df_friday = trips_df[trips_df['service_id'].isin(service_ids_friday)]
trips_df_saturday = trips_df[trips_df['service_id'].isin(service_ids_saturday)]
trips_df_sunday = trips_df[trips_df['service_id'].isin(service_ids_sunday)]

stop_times_df = pd.read_csv('hvv_rohdaten_gtfs/stop_times.txt')

stop_times_df_monday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_monday['trip_id'].values)]
stop_times_df_tuesday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_tuesday['trip_id'].values)]
stop_times_df_wednesday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_wednesday['trip_id'].values)]
stop_times_df_thursday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_thursday['trip_id'].values)]
stop_times_df_friday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_friday['trip_id'].values)]
stop_times_df_saturday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_saturday['trip_id'].values)]
stop_times_df_sunday = stop_times_df[stop_times_df['trip_id'].isin(trips_df_sunday['trip_id'].values)]

days_stop_times_df = [
    stop_times_df_monday,
    stop_times_df_tuesday,
    stop_times_df_wednesday,
    stop_times_df_thursday,
    stop_times_df_friday,
    stop_times_df_saturday,
    stop_times_df_sunday
]

stop_times_datetimes_monday = []
stop_times_datetimes_tuesday = []
stop_times_datetimes_wednesday = []
stop_times_datetimes_thursday = []
stop_times_datetimes_friday = []
stop_times_datetimes_saturday = []
stop_times_datetimes_sunday = []

days_datetimes = [
    stop_times_datetimes_monday,
    stop_times_datetimes_tuesday,
    stop_times_datetimes_wednesday,
    stop_times_datetimes_thursday,
    stop_times_datetimes_friday,
    stop_times_datetimes_saturday,
    stop_times_datetimes_sunday
]

for i in range(len(days_stop_times_df)):
    for index, row in days_stop_times_df[i].iterrows():
        arrival_time = row['arrival_time']
        if arrival_time.startswith('24'):
            arrival_time = arrival_time.replace('24', '00', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        elif arrival_time.startswith('25'):
            arrival_time = arrival_time.replace('25', '01', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        elif arrival_time.startswith('26'):
            arrival_time = arrival_time.replace('26', '02', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        elif arrival_time.startswith('27'):
            arrival_time = arrival_time.replace('27', '03', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        elif arrival_time.startswith('28'):
            arrival_time = arrival_time.replace('28', '04', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        elif arrival_time.startswith('29'):
            arrival_time = arrival_time.replace('29', '05', 1)
            days_datetimes[(i + 1) % 7].append(arrival_time)
        else:
            days_datetimes[i].append(arrival_time)
    
for i in range(len(days_datetimes)):
    with open (f'public_transport_stop_times/{calendar_df.columns[i+1]}' + '.json', 'w') as f:
        json.dump(days_datetimes[i], f)
    """ fig, ax = plt.subplots(figsize=(20, 10))
    ax.hist(days_datetimes[i], bins=24)
    ax.set_title(calendar_df.columns[i+1])
    ax.set_xlabel('Time')
    ax.set_ylabel('Frequency')
    ax.set_xticks(pd.date_range('00:00', '23:59', freq='1H'))
    ax.xaxis.set_tick_params(rotation=90)
    plt.show() """

        
    
```
