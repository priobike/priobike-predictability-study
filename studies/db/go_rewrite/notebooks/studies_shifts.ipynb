```python
import json
import matplotlib.pyplot as plt

with open('processed_things_2024_12_11.json') as f:
    processed_things = json.load(f)
    
shift_sum_zero_counts = []
median_shifts_deviations_from_zero = []

for thing_name, thing in processed_things.items():
    for day in range(7):
        for hour in range(24):
            if thing['ShiftsSumZeroCount'][day][hour] != -999999:
                shift_sum_zero_counts.append(thing['ShiftsSumZeroCount'][day][hour])
            if thing['MedianShiftsDeviationZero'][day][hour] != -999999:
                median_shifts_deviations_from_zero.append(thing['MedianShiftsDeviationZero'][day][hour])
                
fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(shift_sum_zero_counts, bins=100)
ax.set_title('ShiftsSumZeroCount')
plt.show()

fig, ax = plt.subplots(figsize=(10, 10))
ax.hist(median_shifts_deviations_from_zero, bins=100)
ax.set_title('MedianShiftsDeviationZero')
plt.show()
```
